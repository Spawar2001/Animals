```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def detect_anomaly(data_array, k=2, threshold=1000):
    anomalies = []
    for i, point in enumerate(data_array):
        distances = np.abs(data_array - point)
        distances.sort()
        k_nearest = distances[1:k+1]
        mean_distance = np.mean(k_nearest)
        if mean_distance > threshold:
            anomalies.append(point)
    return anomalies
```

```python
df = pd.read_csv('/content/dairy_dataset.csv')
```

```python
Revenue = df['Total Value'].values
```

```python
anomalies = detect_anomaly(Revenue , k=2, threshold=1000)
normal_points = [point for point in Revenue  if point not in anomalies]
```

```python
plt.figure(figsize=(10, 6))
plt.title("Anomoly Detection")
plt.plot(normal_points, np.zeros_like(normal_points), 'o', label='Normal Points')
plt.plot(anomalies, np.zeros_like(anomalies), 'o', color='red', label='Anomalies')
plt.legend()
plt.show()
```
