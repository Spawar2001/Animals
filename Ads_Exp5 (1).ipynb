```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_load = pd.read_csv("/content/dairy_dataset.csv")
data = data_load.iloc[1:500,:]
```

**Scatter_plot**

```python
f = plt.figure()
f.set_figwidth(10)

plt.scatter(data['Product Name'],data['Quantity Sold (liters/kg)'])
plt.title('Product Name VS Quantity Sold  [Scatter_plot]')
plt.xlabel("Product Name")
plt.ylabel("Quantity Sold ")
plt.savefig("fig1.png")
plt.show()
```

**Bar_plot**

```python
f = plt.figure()
f.set_figwidth(15)
plt.bar(data['Brand'],data['Price per Unit'])
plt.title('Brand VS Price per Unit [Bar_plot]')
plt.xlabel("Brand")
plt.ylabel("Price per Unit")
plt.savefig("fig2.png")
plt.show()
```

**Histogram**

```python
f = plt.figure()
f.set_figwidth(10)

import seaborn as sns
axis = sns.histplot(data["Product Name"], kde=True)
plt.title('Histogram')
plt.savefig("fig3.png")
plt.show()
```

**Pie-chart**

```python
f = plt.figure()
f.set_figwidth(10)

data_x = data_load.iloc[6:16,15]
labels = data_load.iloc[6:16,0]
plt.pie(data_x,labels=labels)
# plt.legend(labels,title="Yields",loc="right")
plt.title('QUANTITY-SOLD lOCATION WISE Pie-chart')
plt.savefig("fig4.png")
plt.show()
```

**Box-plot**

```python
temp = data_load.iloc[1:10,:]
plt.boxplot(temp['Quantity in Stock (liters/kg)'])
plt.title('Quantity in Stock  [Box_plot]')
plt.xlabel("Quantity in Stock")
plt.ylabel("frequency")
plt.show()
```

**line-graph**

```python
import seaborn as sn
sn.lineplot(x="Total Land Area (acres)", y="Number of Cows", data=temp)
plt.title('LAND Vs COWS [line Graph]')
plt.savefig("fig5.png")
plt.show()
```

**Heatmap**

```python
f = plt.figure()
f.set_figwidth(10)

import seaborn as sb

dataplot = sb.heatmap(data.corr(), cmap="PiYG", annot=True)
plt.title('Heatmap_corr')
plt.show()
```
